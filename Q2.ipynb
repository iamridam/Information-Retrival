```python
import os
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
import numpy as np
```

```python
yourpath = './20_newsgroups/Q2'
all_files = []
for root, dirs, files in os.walk(yourpath, topdown=False):
    for name in files:
        all_files.append(os.path.join(root, name))
Dict1={}
for i in range(len(all_files)):
    index=0
    file = open(all_files[i], "r") 
    text = file.readlines()
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    stop_words = set(stopwords.words('english'))
    for s in text:
        s_sep = ""
        # Removing Punctuations
        for character in s:
            if character in punctuations:
                outputCharacter = " %s " % character
            else:
                outputCharacter = character
            s_sep += outputCharacter
        # Tokenize sentences to word tokens
        for wrd in TreebankWordTokenizer().tokenize(s_sep):
            # Removed stop words and punctations
            if wrd not in punctuations:
                # Converted all the words to lower case
                wrd_lower = wrd.lower()
                #wrd_lower = lancaster.stem(wrd_lower)
                if ((wrd_lower not in stop_words) & (not wrd_lower.isdigit()) & (len(wrd_lower) > 2)):
                    if wrd_lower in Dict1.keys():
                        if os.path.basename(all_files[i]) in Dict1[wrd_lower].keys():
                            Dict1[wrd_lower][os.path.basename(all_files[i])].append(index)
                        else:
                            Dict1[wrd_lower][os.path.basename(all_files[i])]=[]
                            Dict1[wrd_lower][os.path.basename(all_files[i])].append(index)
                    else:
                        Dict1[wrd_lower]={}
                        Dict1[wrd_lower][os.path.basename(all_files[i])]=[]
                        Dict1[wrd_lower][os.path.basename(all_files[i])].append(index)
            index+=1
```

```python
Query = input("Provide Your Query ")
Q_words = Query.split(" ")
ListOfDocDict = []
for i in Q_words:
    ListOfDocDict.append(Dict1[i])
CommonDocumentsWithoutIndex = list(set(ListOfDocDict[0].keys()) & set(ListOfDocDict[1].keys()))
for i in range(len(ListOfDocDict)-2):
    CommonDocumentsWithoutIndex = list(set(CommonDocumentsWithoutIndex) & set(ListOfDocDict[i+2].keys()))
FinalDocs=[]
for i in CommonDocumentsWithoutIndex:
    Temp = list(np.array(Dict1[Q_words[0]][i])+1)
    for j in range(1,len(Q_words)):
        Temp = list(set(Temp) & set(Dict1[Q_words[j]][i]))
        if(len(Temp)==0):
            break
        Temp = list(np.array(Temp)+1)
    if(len(Temp)!=0):
        FinalDocs.append(i)
print(FinalDocs)
```
